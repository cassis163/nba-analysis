```python
from nba_api.stats.static import players as nba_players

def get_selected_player_names():
    return [line.rstrip("\n") for line in open("./data/players.txt").readlines()]

def get_player(full_name):
    for player in nba_players.get_players():
        if player["full_name"] == full_name:
            return player

def get_selected_players():
    selected_players = []
    for player_name in get_selected_player_names():
        selected_players.append(get_player(player_name))
    
    return selected_players

player = get_player("Wilt Chamberlain")
player
```

```python
from nba_api.stats.endpoints import playercareerstats

stats = ["MIN", "FGM", "REB", "AST", "STL", "BLK", "TOV", "PF", "PTS"]

def add_per_game_stats(df):
    for stat in stats:
        df[stat + "_PG"] = df[stat] / df["GP"]

def add_per_minute_stats(df):
    for stat in stats:
        df[stat + "_PM"] = df[stat + "_PG"] / df["MIN_PG"]

career = playercareerstats.PlayerCareerStats(player_id=player["id"])
df = career.get_data_frames()[0]
add_per_game_stats(df)
add_per_minute_stats(df)
df
```
